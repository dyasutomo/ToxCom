# Word Cloud

```python
from wordcloud import WordCloud, STOPWORDS

def PlotWordCloud(df, col):
    stopwords_ = set(STOPWORDS)
    wordcloud = WordCloud(background_color='white',
                          max_words=1000,
                          stopwords=stopwords_).generate(" ".join([com for com in df[col]]))

    plt.figure(figsize=(16,10),edgecolor='blue')
    plt.imshow(wordcloud, cmap='plasma')
    plt.axis("off")

PlotWordCloud(df[:10000], 'comment_text')
```

### Most used words in the Toxic comments

```python
df['comment_lower'] = df['comment_text'].apply(lambda x: x.lower())
```

```python
from collections import Counter
STOPWORDS = set(stopwords.words('english'))
words_in_toxic_coms = [word for com in df[df['label_sum']!=0]['comment_lower'].values for
                       word in com.lower().split() if word not in STOPWORDS ]
```

```python
def plot_common_words(words_list, topN=100, ax=None):    
    word_counter=Counter(words_list)
    most_common = word_counter.most_common(topN)
    ks = [kv_pair[0] for kv_pair in most_common]
    vs = [kv_pair[1] for kv_pair in most_common]

    print ("Top 10 most common words:", most_common[:10] )
    
    if ax is None:
        fig, ax = plt.subplots(1,1,figsize=(20,4));
    
    sns.barplot(x=ks, y=vs, ax=ax);
    ax.tick_params(axis='x', which='major', labelsize=14, rotation=90)
    ax.tick_params(axis='y', which='major', labelsize=20)
    
plot_common_words(words_in_toxic_coms)
```

There are quiet a lot of repetition of words like 'F\*\*k' and 'F\*\*king'. So, let's stemmize the words.

```python
from nltk.stem.porter import PorterStemmer

p_stemmer = PorterStemmer()
words_in_toxic_coms_stem = [p_stemmer.stem(word) for com in df[df['label_sum']!=0]['comment_lower'].values for word in com.lower().split() if word not in STOPWORDS ]
plot_common_words(words_in_toxic_coms_stem)
```

```python

fig, ax = plt.subplots(2,1,figsize=(20,10));
ax[0].set_title("Top Words before Stemming", fontsize=25)
plot_common_words(words_in_toxic_coms, topN=50, ax=ax[0])


ax[1].set_title("Top Words after Stemming", fontsize=25)
plot_common_words(words_in_toxic_coms_stem, topN=50, ax=ax[1])

plt.subplots_adjust(hspace=0.5)
```

## Text preprocessing

```python
def prepare_text(text):
    text = text.lower()
    # remove \n
    text = re.sub("\\n", " ", text)
    # remove 've, 're
    text = re.sub("[a-z]*\'[r,v]e", "", text) 
    # remove 's, 't, 'r, 'v
    text = re.sub("[a-z]*\'[s,t,r,v]", "", text) 
    # Replace everything not a letter with a space
    text = re.sub("[^a-zA-Z]", " ", text)
    # Remove stopwords
    text = ' '.join([word for word in text.split() if word not in STOPWORDS])
    return text
```

```python
df['com_processed'] = df['comment_text'].apply(prepare_text)
df.head(3)
```

```python
def get_token(com):
    com = re.sub("\\n", " ", com)
    return com.split()

df['tokens'] = df['comment_lower'].apply(get_token)
df['tokens_num'] = df['tokens'].apply(len)
df.head(2)
```

```python
df_bin0=df[df['bin_label']==0]
df_bin1=df[df['bin_label']==1]

token_0 = int(df_bin0['tokens_num'].sum()/df_bin0.shape[0])
token_1 = int(df_bin1['tokens_num'].sum()/df_bin1.shape[0])

apos_0=np.round(df_bin0['apos_tokens_len'].sum()/df_bin0.shape[0], 1)
apos_1=np.round(df_bin1['apos_tokens_len'].sum()/df_bin1.shape[0], 1)

apo_tok0=np.round(100*apos_0/token_0,1)
apo_tok1=np.round(100*apos_1/token_1,1)

data = [ [token_0, token_1],
         [apos_0, apos_1],
         [apo_tok0, apo_tok1]]

titles=["Avg. Tokens \nPer Comment",
        "Avg. Apostrophe Tokens \nPer Comment",
        "Weighted Avg. Apostrophe Tokens \nPer Comment"]

texts = ["", "", " %"]

fig, ax=plt.subplots(1,3, figsize=(15,4))
for i in range(3):
    d0,d1=data[i]
    ax[i].bar(['Non-Toxic', 'Toxic'],[d0, d1], color=['blue', 'red']);
    ylim=max(d0,d1)*1.15
    ax[i].set_title(titles[i], fontsize="xx-large")
        
    ax[i].text(-.15, 1.02*d0, str(d0)+texts[i], color='blue', fontsize='xx-large');
    ax[i].text(.85, 1.02*d1, str(d1)+texts[i], color='red', fontsize='xx-large');
    ax[i].set_ylim([0, ylim])
    ax[i].set_xticklabels(['Non-Toxic', 'Toxic'],fontsize= 'xx-large');
    ax[i].set_yticks([])
```
